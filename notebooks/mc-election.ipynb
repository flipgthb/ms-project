```python
def save_data(path, data_tuple, time_stamp=True):

    ts = ""
    if time_stamp:
        lt = time.localtime()
        ymd = lt.tm_year,lt.tm_mon,lt.tm_mday
        date = datetime.date(*ymd)
        ts = "_"+str(date)


    data = dict.fromkeys(data_tuple[0].keys())
    for entry in data_tuple:
        for k,v in entry.items():
            try:
                data[k].append(v)
            except:
                data[k] = [v]

    final = {}
    for file_name, data_list in data.items():
        full_name = path + file_name + ts + ".csv"
        with open(full_name, "w") as file_:
            panel = pd.concat(data_list)
            panel.to_frame().to_csv(file_)
            final[file_name] = panel
    return final      

def get_panel(path, header=[0,1,2], index_col=[0,1]):
    with open(path) as file_:
        dataframe = pd.read_csv(file_, header=header, index_col=index_col)
        panel = dataframe.to_panel()
        lvl = list(panel.items.levels)
        new_lvl = map(lambda x: x.astype(float), lvl)
        panel.items.set_levels(new_lvl, inplace=True)
    return panel
```

```python
%%px
from __future__ import division
import numpy as np
import pandas as pd
import networkx as nx
import math
from scipy.special import erfc

sqrt2 = math.sqrt(2)
sqrt2pi = math.sqrt(2*np.pi)
sqrtK = math.sqrt(K)

def montecarlo(eps,rho,beta):
    
    n1,n2,n3 = party_sizes
    N = sum(party_sizes)
    gm = math.sqrt(1-rho**2)/rho
    Jg, Jo = political_pressure
    #rho1, rho2, rho3 = 3*[rho]
    
    agenda1 = [-1,-1,-1,-1,-1.0]
    #agenda1 = [3/2,-1/2,1,-1,-1.0]
    agenda2 = [1,1,1,1,1.0]
    
    agenda1 = sqrtK*np.array(agenda1)/np.linalg.norm(agenda1)
    agenda2 = sqrtK*np.array(agenda2)/np.linalg.norm(agenda2)
    
    w1 = np.vstack(n1*[agenda1]) + 0.1*np.random.randn(n1, K)
    w2 = np.vstack(n2*[agenda2]) + 0.1*np.random.randn(n2, K)
    w3 = np.random.randn(n3, K)# + 0.1*np.random.randn(n3, K)

    w = np.vstack([w1,w2,w3])
    w *= sqrtK/np.linalg.norm(w,axis=1)[:,None]
    agenda1 *= sqrtK/np.linalg.norm(agenda1)
    agenda2 *= sqrtK/np.linalg.norm(agenda2)
    agendas = np.vstack([agenda1, agenda2])
    h = agendas.dot(w.T)/sqrtK
    
    O = lambda n, m: np.zeros((n,m))
    I = lambda n, m: np.ones((n,m))

    A1 = np.hstack([I(n1,n1), O(n1, n2), O(n1, n3)])
    A2 = np.hstack([O(n2,n1), I(n2, n2), O(n2, n3)])
    A3 = np.hstack([I(n3,n1), I(n3, n2), I(n3, n3)])
    A = np.vstack([A1, A2, A3])

    party1 = slice(0, n1)
    party2 = slice(n1, n1+n2)
    party3 = slice(n1+n2, n1+n2+n3)
    J = np.array(n1*[Jg]+n2*[Jo]+n3*[Jo])
    L = np.array(n1*[0]+n2*[1]+n3*[-1])
    
    def measure():
        m = h.mean(axis=1)/sqrtK
        a31 = h[0, party3].mean()/sqrtK
        a32 = h[1, party3].mean()/sqrtK
        a11 = h[0, party1].mean()/sqrtK
        a22 = h[1, party2].mean()/sqrtK
        v31 = (h[0,party3]*h[0,party3][:,None]/K).mean()
        v32 = (h[1,party3]*h[1,party3][:,None]/K).mean()
        data = pd.Series([m[0],m[1], a31, a32, a11, a22, v31, v32], index="m1 m2 a31 a32 a11 a22 v31 v32".split(" "))
        return data
    
    trace = pd.DataFrame()
    got = measure()
    got["acceptance_ratio"] = 0
    trace = trace.append(got, ignore_index=True)
    
    def energy(hi, hj):
        z = hi*np.sign(hj)/gm
        Jij = 1/K
        v = -Jij*gm*gm*math.log(eps + (1-2*eps)*erfc(-z/sqrt2)/2)
        return v
    
    accepted = 0
    for t in xrange(M*N):
        i = np.random.choice(N)
        ni = A[i]
        pij = ni/ni.sum()
        j = np.random.choice(N, p=pij)
        
        Jp = J[j]
        k = L[j]
        if k==-1:
            k = np.random.choice([0,1])
            
        hi, hj = h[k,i], h[k,j]
        
        E0 = energy(hi,hj)
        u = np.random.multivariate_normal(w[i], dw*np.eye(K))
        u *= sqrtK / np.linalg.norm(u)
        l = agendas[k].dot(u)/sqrtK
        E = energy(l, hj)
        if E < E0 or np.random.rand() < math.exp(Jp*beta*(E0-E)):
            accepted += 1
            w[i] = u.copy()
            h[k,i] = l
        
        if t%N == 0:
            got = measure()
            got["acceptance_ratio"] = accepted/(t+1)
            trace = trace.append(got, ignore_index=True)
            
        if t==N*M/2:
            J = np.array(n1*[Jo]+n2*[Jg]+n3*[Jo])
            
    p = (eps, rho, beta)
    names = "epsilon rho beta".split(' ')
    entries = dict(statistics=trace,
                   agents=pd.DataFrame(w),
                   network=pd.DataFrame(A))
    data = {}
    for k,v in entries.items():
        dv = pd.Panel({p:v})
        dv.items.set_names(names, inplace=True)
        data[k] = dv
    return data
```
