```python
def save_data(path, data_tuple, time_stamp=True):

    ts = ""
    if time_stamp:
        lt = time.localtime()
        ymd = lt.tm_year,lt.tm_mon,lt.tm_mday
        date = datetime.date(*ymd)
        ts = "_"+str(date)


    data = dict.fromkeys(data_tuple[0].keys())
    for entry in data_tuple:
        for k,v in entry.items():
            try:
                data[k].append(v)
            except:
                data[k] = [v]

    final = {}
    for file_name, data_list in data.items():
        full_name = path + file_name + ts + ".csv"
        with open(full_name, "w") as file_:
            panel = pd.concat(data_list)
            panel.to_frame().to_csv(file_)
            final[file_name] = panel
    return final      

def get_panel(path, header=[0,1,2], index_col=[0,1]):
    with open(path) as file_:
        dataframe = pd.read_csv(file_, header=header, index_col=index_col)
        panel = dataframe.to_panel()
        lvl = list(panel.items.levels)
        new_lvl = map(lambda x: x.astype(float), lvl)
        panel.items.set_levels(new_lvl, inplace=True)
    return panel
```

```python
%%px
from __future__ import division
import numpy as np
import pandas as pd
import networkx as nx
import math
from scipy.special import erfc

sqrt2 = math.sqrt(2)
sqrt2pi = math.sqrt(2*np.pi)
sqrtK = math.sqrt(K)

def montecarlo(eps,rho,beta):
    
    gm = math.sqrt(1-rho**2)/rho
    
    w = np.random.randn(N,K)
    w *= sqrtK / np.linalg.norm(w, axis=1)[:,None]
    
    #x =  sqrtK * np.array(zeitgeist) / np.linalg.norm(zeitgeist)
    
    #h = x.dot(w.T)/sqrtK
    
    G = nx.complete_graph(N)
    #G = nx.random_regular_graph(20, N)
    A0 = np.asarray(nx.attr_matrix(G)[0])
    A0 *= np.random.rand(*A0.shape)
    #A0 = np.asarray(nx.attr_matrix(G)[0])*3
    #A0 = np.random.rand(N,N)
    #A0 -= np.diag(np.diag(A0))
    A = A0.copy()
    
    def measure():
        #m = h.mean()/sqrtK
        #r = np.abs(h).mean()/sqrtK
        #v = h.var()/K
        #q = (A*h*h[:,None]).mean()
        Q = np.linalg.norm(w.mean(axis=0))
        q = np.linalg.norm(w, axis=1).mean()
        enemies = np.where(A<1/2,1,0)
        n = enemies.mean()
        data = pd.Series([Q,q,n], index="Q q n".split(" "))
        return data
    
    trace = pd.DataFrame()
    got = measure()
    got["acceptance_ratio"] = 0
    trace = trace.append(got, ignore_index=True)
    
    def energy(hi,hj,rij):
        z = hi*np.sign(hj)/gm
        #z = hi*hj/gm
        #Jij = np.sign(hi*hj)/K
        Jij = 1/K
        #Jij = np.sign(rij-1/2)/K
        #Jij = (rij - 1/2)/K
        v = -Jij*gm*gm*math.log(eps + (1-2*eps)*erfc(-z/sqrt2)/2)
        return v
    
    accepted = 0
    for t in xrange(M*N):
        i = np.random.choice(N)
        ni = A[i]
        #ni = np.exp(N*A[i])
        #ni = np.where(A[i] <= 1e-3, 1, 0)
        #ni[i] = 0
        #pij = ni/ni.sum()
        pij = ni/ni.sum()
        j = np.random.choice(N, p=pij)
        
        x = np.random.multivariate_normal(w[i], 0.3*np.eye(K))
        x *= sqrtK / np.linalg.norm(x)
        hi = x.dot(w[i])/sqrtK
        hj = x.dot(w[j])/sqrtK
        #hi, hj = h[i], h[j]
        rij = A[i,j]
        rij = rij + dr*rij*(1-rij)*np.sign(hi*hj)
        #rij += np.sign(hi*hj)
        A[i,j] = rij
        #A[j,i] = rij
        
        E0 = energy(hi,hj,rij)
        u = np.random.multivariate_normal(w[i], dw*np.eye(K))
        u *= sqrtK / np.linalg.norm(u)
        l = x.dot(u)/sqrtK
        E = energy(l, hj,rij)
        if E < E0 or np.random.rand() < math.exp(beta*(E0-E)):
            accepted += 1
            w[i] = u.copy()
        
        if t%N == 0:
            got = measure()
            got["acceptance_ratio"] = accepted/(t+1)
            trace = trace.append(got, ignore_index=True)
            
    p = (eps, rho, beta)
    names = "epsilon rho beta".split(' ')
    entries = dict(statistics=trace,
                   agents=pd.DataFrame(w),
                   network=pd.DataFrame(A))
    data = {}
    for k,v in entries.items():
        dv = pd.Panel({p:v})
        dv.items.set_names(names, inplace=True)
        data[k] = dv
    return data
```
