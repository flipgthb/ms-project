```python
delta, beta = np.meshgrid(np.arange(0,100,1),
                          np.arange(0,1,.01))

mA1 = M[:,:,0,0]
mB1 = M[:,:,0,1]
mA2 = M[:,:,1,0]
mB2 = M[:,:,1,1]
m = [(mA1, r"grupo A, zeitgeits=1"),
     (mB1, r"grupo B, zeitgeits=1"),
     (mA2, r"grupo A, zeitgeits=2"),
     (mB2, r"grupo B, zeitgeits=2")]

f = pl.figure(figsize=(9,7))
ax = []
im = []
cb = []
for i in xrange(4):
    axes = f.add_subplot(2,2,i)
    img = axes.pcolormesh(delta, beta, m[i][0], cmap=pl.cm.RdBu)
    c = f.colorbar(img, ax=axes)
    axes.set_xlabel(r"$\beta$")
    axes.set_ylabel(r"$\delta$")
    axes.set_title(m[i][1])
    ax.append(axes)
    im.append(img)
    cb.append(c)
f.tight_layout()
```

```python
ms = [((mA1-mA2)/2, r"grupo A - ms"),
     ((mB1-mB2)/2, r"grupo B - ms"),
     ((mA1+mA2)/2, r"grupo A - m"),
     ((mB1+mB2)/2, r"grupo B - m")]

f2 = pl.figure(figsize=(9,7))
ax2 = []
im2 = []
cb2 = []
for i in xrange(4):
    axes = f2.add_subplot(2,2,i)
    img = axes.pcolormesh(delta, beta, ms[i][0], cmap=pl.cm.RdBu)
    c = f2.colorbar(img, ax=axes)
    axes.set_xlabel(r"$\beta$")
    axes.set_ylabel(r"$\delta$")
    axes.set_title(ms[i][1])
    ax2.append(axes)
    im2.append(img)
    cb2.append(c)
f2.tight_layout()
```
