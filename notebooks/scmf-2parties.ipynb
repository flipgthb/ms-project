```python
from __future__ import division, print_function
import numpy as np
from scipy.integrate import quad
from scipy.optimize import fixed_point
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from IPython.display import display
from IPython.html import widgets
sns.set_style("white")
sns.set_style({"font.family":"Ubuntu", "axes.grid": False,
               "grid.color": "0.85", "font.size":20})
sns.set_context("notebook", {"lines.linewidth": 2.0,
                             "figure.figsize":np.array([12,10])})
sns.set_palette("Set1")
%matplotlib inline
```

```python
def Psi1(h, m1, r1, m2, r2, n1, n2, rho, tau, alpha, beta):
    A = (1 - rho)/2
    B = rho/2
    Vmf1 = -A*h*m1 + B*np.abs(h)*r1 - np.abs(tau-h*m1)/2 - tau/2
    Vmf2 = -A*h*m2 + B*np.abs(h)*r2 - np.abs(tau-h*m2)/2 - tau/2
    psi = (1 - h*h)*np.exp(-beta*(Vmf1 + alpha*Vmf2))
    return psi

def Psi2(h, m1, r1, m2, r2, n1, n2, rho, tau, alpha, beta):
    A = (1 - rho)/2
    B = rho/2
    Vmf1 = -A*h*m1 + B*np.abs(h)*r1 - np.abs(tau-h*m1)/2 - tau/2
    Vmf2 = -A*h*m2 + B*np.abs(h)*r2 - np.abs(tau-h*m2)/2 - tau/2
    psi = (1 - h*h)*np.exp(-beta*(Vmf2 + alpha*Vmf1))
    return psi

def Z1(*args):
    z = quad(Psi1, -1, 1, args=args)[0]
    return z

def Z2(*args):
    z = quad(Psi2, -1, 1, args=args)[0]
    return z

def m1(z, *args):
    f = lambda x: x*Psi1(x, *args)
    m = quad(f, -1, 1)[0]/z
    return m

def m2(z, *args):
    f = lambda x: x*Psi2(x, *args)
    m = quad(f, -1, 1)[0]/z
    return m

def r1(z, *args):
    f = lambda x: np.abs(x)*Psi1(x,*args)
    r = quad(f, -1, 1)[0]/z
    return r

def r2(z, *args):
    f = lambda x: np.abs(x)*Psi2(x,*args)
    r = quad(f, -1, 1)[0]/z
    return r

def sce(m10, r10, m20, r20, n1, n2, rho, eps, alpha, beta):
    n = 0
    converged = False
    M1 = m10
    R1 = r10
    M2 = m20
    R2 = r20
    delta = .1
    g = np.sqrt(1-rho**2)/rho
    tau = -g + g*np.log(2*eps)/2
    while (not converged) and n < 500:
        M10 = M1
        R10 = R1
        M20 = M2
        R20 = R2
        z1 = Z1(M10, R10, M20, R20, n1, n2, rho, tau, alpha, beta)
        z2 = Z2(M10, R10, M20, R20, n1, n2, rho, tau, alpha, beta)
        M1 = (1 - delta)*M10 + delta*m1(z1, M10, R10, M20, R20, n1, n2, rho, tau, alpha, beta)
        R1 = (1 - delta)*R10 + delta*r1(z2, M10, R10, M20, R20, n1, n2, rho, tau, alpha, beta)
        M2 = (1 - delta)*M20 + delta*m2(z1, M10, R10, M20, R20, n1, n2, rho, tau, alpha, beta)
        R2 = (1 - delta)*R20 + delta*r2(z2, M10, R10, M20, R20, n1, n2, rho, tau, alpha, beta)
        if (np.abs(M1 - M10)<1e-5 or np.abs(R1 - R10)<1e-5) and (np.abs(M2 - M20)<1e-5 or np.abs(R2 - R20)<1e-5):
            converged = True 
        elif n == 499:
            print("Failed to converge!")
        n += 1
    return pd.Series([M1, R1, M2, R2, (M1-M2)/2, (M1+M2)/2, converged, n],
                     index=['m_1','r_1','m_2','r_2', 'm_s', 'm', 'converged', 'n_iter'],
                     name=(m10, r10, m20, r20, n1, n2, rho, eps, alpha, beta))
```

```python
def get_float_slider(list_, trim=2):
    x0, x = min(list_), max(list_)
    n = len(list_)-1
    if n != 0:
        dx = (x-x0)/n
    else:
        dx = 0.0
    s = [round(a, trim) for a in (x0, x, dx)]
    sld = widgets.FloatSliderWidget(min=s[0], max=s[1], 
                                    step=s[2], value=s[0])
    return sld

def plot_data(m10,r10,n1,rho,epsilon,alpha):
    m20 = -m10
    r20 = r10
    n2 = 1 - n1
    m1_data = data['m_1'][m10][r10][n1][rho][epsilon][alpha]
    r1_data = data['r_1'][m10][r10][n1][rho][epsilon][alpha]
    m2_data = data['m_2'][m10][r10][n1][rho][epsilon][alpha]
    r2_data = data['r_2'][m10][r10][n1][rho][epsilon][alpha]
    ms_data = data['m_s'][m10][r10][n1][rho][epsilon][alpha]
    m_data =  data['m']  [m10][r10][n1][rho][epsilon][alpha]
    
    f = plt.figure(figsize=(10, 16))
    ax1 = f.add_subplot(3,1,1)
    m1_data.plot(ax=ax1, label=r"$m_1$")
    m2_data.plot(ax=ax1, label=r"$m_2$")
    ax1.set_xlabel(r"$\beta$", fontdict={"size":30})
    ax1.legend(loc=(1.05, 0), prop={"size":30})
    
    ax2 = f.add_subplot(3,1,2)
    r1_data.plot(ax=ax2, label=r"$r_1$", alpha=.5)
    r2_data.plot(ax=ax2, label=r"$r_2$", alpha=.5)
    ax2.set_xlabel(r"$\beta$", fontdict={"size":30})
    ax2.legend(loc=(1.05, 0), prop={"size":30})
    
    ax3 = f.add_subplot(3,1,3)
    ms_data.plot(ax=ax3, label=r"$m_s$")
    m_data.plot(ax=ax3, label=r"$m$")
    ax3.set_xlabel(r"$\beta$", fontdict={"size":30})
    ax3.legend(loc=(1.05, 0), prop={"size":30})
    
    sns.despine(f)
    
def plt_wgt():
    prmt = ['m10','r10','n1','rho','epsilon','alpha']
    vl = [get_float_slider(x.tolist()) for x in data.index.levels]
    args = dict(zip(prmt, vl))
    wgt = widgets.interactive(plot_data, **args)
    return wgt
```
