```python
%matplotlib inline
from __future__ import division
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = lambda x,l,s: x + l*x*(1-x)*s
x0 = 1/2
nmax = 500
t = 0
l = 1.0
cmax = 20
nplus, nminus = cmax/2, cmax/2
ps = [nminus/cmax,nplus/cmax]
s = np.random.choice([-1,1], p=ps)
hp, hm = 0, 0
if s > 0:
    hp += 1
else:
    hm += 1
x = f(x0, l, s)
err = abs(x-x0)
while err > 1e-5 and t < nmax:
    print("x0={}, x={}, s={}, ps={},err={}, t={}, hp={}, hm={}".format(x0,x,s,ps,err,t,hp,hm))
    x0 = x
    s = np.random.choice([-1,1], p=ps)
    if s > 0:
        nplus += 1
        nminus -= 1
        hp += 1
    else:
        nminus += 1
        nplus -= 1
        hm += 1
    nplus += 1
    nminus -= 1
    ps = [nminus/cmax,nplus/cmax]
    x = f(x0, l, s)
    err = abs(x-x0)
    t += 1
print("x0={}, x={}, s={}, ps={},err={}, t={}, hp={}, hm={}".format(x0,x,s,ps,err,t,hp,hm))
```

```python
from scipy.special import erfc
def H(x):
    return erfc(x/np.sqrt(2))/2

def V(x, r, e, K):
    g = np.sqrt(1-r*r)/r
    v = -g*g*np.log(e + (1-2*e)*H(-x/g))/K
    return v

x = np.arange(-3,3,0.0001)
e, r, K = 0.2, 0.5, 5
plt.plot(x,-np.gradient(V(x, r, e, K)))
```

```python
def F(x, r, e, K):
    g = np.sqrt(1-r*r)/r
    f = g*(1-2*e)*np.exp(-x*x/g/g/2)/(e + (1-2*e)*H(-x/g))/np.sqrt(2*np.pi)/K
    return f

plt.plot(x,F(x,r,e,K))
plt.vlines(x[np.abs(F(x,r,e,K)-F(x,r,e,K).max()/2)<1e-4], 0, 0.2)
```

```python
def Fmf(x,r,e,K):
    g = np.sqrt(1-r*r)/r
    t = g*np.log(2*e)
    f = (1-r)/2 - r*np.sign(x)/2 + np.sign(x+t)/2
    return f

plt.plot(x, Fmf(x,r,e,K))
```
