```python

sim = "simulation-3/"
path = base_path + sim

statistics = pd.read_csv(path+"statistics.csv")
cumulative = pd.read_csv(path+"cumulative.csv")
social_network = np.load(path+"social-network.npy")
state = np.load(path+"state.npy")
zeitgeist = np.load(path+"zeitgeist.npy")

R_max = statistics["R_max"].reshape((6,100))
R_mean = statistics["R_mean"].reshape((6,100))
A_max = statistics["A_max"].reshape((6,100))
A_mean = statistics["A_mean"].reshape((6,100))
delta = statistics["delta"].reshape((6,100))
beta = statistics["beta"].reshape((6,100))
m = statistics["m"].reshape((6,100))
print(delta.shape, beta.shape, R_max.shape, R_mean.shape)

c = ['r','c','b','y','g','m']
f = pl.figure(figsize=(9,7))
ax = []
im = []
for i in xrange(6):
    axes = f.add_subplot(2,3,i)
    img = axes.plot(beta[i], R_mean[i]/R_max[i], '%s-'%c[i])
    #img = axes.plot(beta[i], A_mean[i]/A_max[i], '%s-'%c[i])
    #img = axes.plot(beta[i], np.abs(m[i]), '%s-'%c[i])
    axes.set_xlabel(r"$\beta$")
    axes.set_ylabel(r"$\frac{<R>}{R_{max}}$")
    axes.set_title(r"$\delta=%.1f$"%delta[i,0])
    ax.append(axes)
    im.append(img)
f.tight_layout()
#print(R_max[5])
```
