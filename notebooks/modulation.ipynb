```python
%matplotlib inline
from __future__ import division, print_function
import numpy as np
import matplotlib.pyplot as plt
import math
from scipy.special import erf, erfc
import seaborn as sbn

sbn.set_style("white")
sbn.set_style("ticks")
sbn.set_style({"font.family":"Ubuntu", "axes.grid": False,
               "grid.color": "0.85"})
sbn.set_context("notebook", {"lines.linewidth": 2.0})
sbn.set_palette("Set1")
```

```python
def H(x):
    return erfc(x/math.sqrt(2))/2

def E(z, rho, eta):
    Gamma = math.sqrt(1-rho*rho)/rho
    v = -Gamma*np.log(eta + (1-2*eta)*H(-z/Gamma))
    return v

def F(z, rho, eta):
    Gamma = np.sqrt(1 - rho*rho)/rho
    x = z/Gamma
    num = (1-2*eta)*np.exp(-x*x/2)/math.sqrt(2*np.pi)
    den = eta + (1-2*eta)*H(-z/Gamma)
    v = num/den
    return v

f = plt.figure(figsize=(14,10))
ax3 = f.add_subplot(1,1,1)
eta = .15
hs = np.linspace(-5, 5,1000)
for rho in np.arange(.5, .91, .1):
    Fhs = -np.gradient(E(hs, rho, eta))
    #Fhs = F(hs, rho, eta)
    #max_hs = Fhs[Fhs==Fhs.max()][0]
    l, = ax3.plot(hs, Fhs, 
                  label=r"$\rho=%.1f$"%rho,#, $x_{max}=%.2f$"%(rho,max_hs),
                  alpha=.6)
    #ax3.vlines(max_x, 0, f.max(),
    #           colors=l.get_color(), alpha=.6)
ax3.legend(loc=(1.05,0))
ax3.set_ylim(bottom=0)
sbn.despine()
```

```python
def F(hs, rho, eta):
    Gamma = np.sqrt(1 - rho*rho)/rho
    arg = hs/Gamma
    eta= 0
    num = Gamma*Gamma*(1-2*eta)*np.exp(-arg*arg)/math.sqrt(2*np.pi)
    den = eta + (1-2*eta)*H(-arg)
    v = num/den
    return v

HS = np.linspace(-5, 5, 1000)
plt.plot(HS, F(HS, .5, .0))
```
