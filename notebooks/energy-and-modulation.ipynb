```python
from __future__ import division, print_function
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import erfc
import math
import seaborn as sns
sns.set_style("white")

#sns.set_style("ticks")
sns.set_style({"font.family":"Ubuntu", "axes.grid": False,
               "grid.color": "0.85"})
sns.set_context("notebook", {"lines.linewidth": 2.0})
sns.set_palette("Set1")
%matplotlib inline
```

```python
def H(x):
    return erfc(x/math.sqrt(2))/2

def F(z, rho, eta):
    G = np.sqrt(1-rho*rho)/rho
    x = z/G
    num = (1-2*eta)*np.exp(-x*x/2)
    den = eta + (1-2*eta)*H(-x)
    return num/den

def E(z, rho, eta):
    G = np.sqrt(1-rho*rho)/rho
    v = -G*np.sqrt(2*np.pi)*np.log(eta + (1-2*eta)*H(-z/G))
    return v

f = plt.figure(figsize=(16,8))
ax1 = f.add_subplot(1,2,1)
ax2= f.add_subplot(1,2,2)
hs = np.linspace(-6, 6, 1000)
rho = np.arange(.2, .9, .1)
colors = sns.color_palette("Set3", n_colors=rho.shape[0])
eps = .2
for r in rho:
    l1, = ax1.plot(hs, -np.gradient(E(hs, r, eps)), label=r"$\rho=%.1f$"%r, c=colors[np.where(rho==r)[0]])
    #ax1.fill_between(hs, F(hs, r, eps), alpha=.1, color=colors[np.where(rho==r)[0]])
    ax2.plot(hs, E(hs, r, eps), label=r"$\rho=%.1f$"%r, c=colors[np.where(rho==r)[0]])
    ax2.fill_between(hs, E(hs, r, eps), alpha=.1, color=colors[np.where(rho==r)[0]])
ax2.legend(loc=(1.05, 0))
ax1.grid()
ax2.grid()
ax1.vlines(ax1.get_xlim()[1], ax1.get_ylim()[0], ax1.get_ylim()[1], lw=2, alpha=.2)
ax2.vlines(ax2.get_xlim()[1], ax2.get_ylim()[0], ax2.get_ylim()[1], lw=2, alpha=.2)
ax1.set_ylim(bottom=0)#, top=10)
ax2.set_ylim(bottom=0)#, top=50)
ax1.spines['left'].set_position("center")
ax2.spines['left'].set_position("center")
ax1.spines['left'].set_linewidth(2)
ax2.spines['left'].set_linewidth(2)
ax1.spines['bottom'].set_linewidth(2)
ax2.spines['bottom'].set_linewidth(2)
ax1.set_yticks(ax1.get_yticks()[1:])
ax2.set_yticks(ax2.get_yticks()[1:])
sns.despine(f)
```

```python
def invF(y, rho, eps):
    g = np.sqrt(1-rho**2)/rho
    fi = -np.sqrt(2*np.pi)/(1-2*eps)/2 + g/y
    return fi

def tau(alpha, rho, eps):
    g = np.sqrt(1-rho**2)/rho
    return -invF(1 - alpha/2, rho, eps)/g

RHO = np.arange(.2, .9, .3)
EPS = np.linspace(0, 1/2, 1000)
for rho in RHO:
    alpha = rho
    plt.plot(EPS, tau(alpha, rho, EPS), label=r"$\rho=%.1f$"%rho)
#plt.gca().set_ylim(bottom=0, top=5)
plt.legend(loc=(1.05,0))
sns.despine()
```
