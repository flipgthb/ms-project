```python
%%px

from __future__ import division, print_function
import numpy as np
from scipy.integrate import quad
import pandas as pd

def Psi(h, m, r, rho, eps, beta):
    psi = (1 - h*h)*np.exp(beta*(m*h - rho*r*np.abs(h)/2 + rho*np.abs(m*h+(1-2*eps))/2))
    return psi

def Z(*args):
    z = quad(Psi, -1, 1, args=args)[0]
    return z

def m(z, *args):
    f = lambda x: x*Psi(x, *args)
    m = quad(f, -1, 1)[0]/z
    return m

def r(z, *args):
    f = lambda x: np.abs(x)*Psi(x,*args)
    r = quad(f, -1, 1)[0]/z
    return r

def sce(m0, r0, rho, eps, betaK):
    n = 0
    converged = False
    M = m0
    R = r0
    g = np.sqrt(1-rho**2)/rho
    tau = (-g + g*np.log(2*eps)/2)/30
    delta = .1
    precision = 1e-6
    if betaK >= 3 and betaK < 7:
        precision = 1e-8
    while (not converged) and n < 10000:
        M0 = M
        R0 = R
        z = Z(M0, R0, rho, tau, betaK)
        M = (1 - delta)*M0 + delta*m(z, M0, R0, rho, tau, betaK)
        R = (1 - delta)*R0 + delta*r(z, M0, R0, rho, tau, betaK)
        if np.abs(M - M0)<precision or np.abs(R - R0)<precision:
        # if np.abs(M - M0)<precision and np.abs(R -R0)<precision:
            converged = True
        n += 1
    data = pd.Series([M, R, converged, n],
                     index=['m', 'r', 'converged', 'n_iter'],
                     name=(m0, r0, rho, eps, betaK))
    return data

def P(h, m, r, rho, eps, beta):
    z = Z(m, r, rho, eps, beta)
    return Psi(h, m, r, rho, eps, beta)/z
```

```python
m0_grid =   np.arange(-.1,.2,.1)
r0_grid =   np.arange(0,.7,.3)
rho_grid =  np.linspace(1e-4,0.9,5)
eps_grid =  np.linspace(1e-4, 0.45, 5)
beta_grid = np.hstack([np.linspace(0,5,10),
                       np.linspace(5,15,50),
                       np.linspace(15,30,10)])
grid = np.meshgrid(m0_grid, r0_grid, rho_grid, eps_grid, beta_grid)
points = np.vstack([x.ravel() for x in grid]).T
dv.scatter('points', points)
```

```python
%%px

def simulation(p):
    m0, r0, r, e, b = p
    M, R = sce(m0, r0, r, e, b)
    res = pd.Series(
             [m0,r0,r,e,b,M,R], 
             index=["m0","r0","rho","epsilon","beta","m","r"]
          )
    return res
```

```python
%%px
import time
from datetime import timedelta
t0 = time.time()
data = map(simulation, points)
t = time.time()
print("time spent: ", timedelta(seconds=(t-t0)))
```

```python
def get_float_slider(list_, trim=2):
    x0, x = min(list_), max(list_)
    n = len(list_)-1
    if n != 0:
        dx = (x-x0)/n
    else:
        dx = 0.0
    s = [round(a, trim) for a in (x0, x, dx)]
    sld = widgets.FloatSliderWidget(min=s[0], max=s[1], 
                                    step=s[2], value=s[0])
    return sld

def plot_data(m0, r0, rho, epsilon):
    m_data = data['m'][m0][r0][rho][epsilon]
    r_data = data['r'][m0][r0][rho][epsilon]
    f = plt.figure(figsize=(10, 10))
    ax = f.add_subplot(1,1,1)
    m_data.plot(ax=ax, label=r"$m$")
    r_data.plot(ax=ax, label=r"$r$")
    ax.set_xlabel(r"$\beta$", fontdict={"size":30})
    ax.legend(loc=(1.05, 0), prop={"size":30})
    sns.despine(f)
    
def plt_wgt():
    prmt = ['m0', 'r0', 'rho', 'epsilon']
    vl = [get_float_slider(x.tolist()) for x in data.index.levels]
    args = dict(zip(prmt, vl))
    wgt = widgets.interactive(plot_data, **args)
    return wgt
```
