```python
def H(x):
    return erfc(x/np.sqrt(2))/2

def F(th, rho, eps):
    gamma = np.sqrt(1-rho**2)/rho
    A = (1-2*eps)/np.sqrt(2*np.pi)
    x = th/gamma
    num = np.exp(-x*x/2)
    den = eps + (1-2*eps)*H(-x)
    return A*num/den
```

```python
def ilustration(ax):
    sns.despine(ax=ax)
    #xmin, xmax = ax.get_xlim()
    #xt = sorted([xmin/2,0,xmax/2])
    ax.set_xticks([0])
    #ax.set_xticklabels(["<-surpresa","0","esperado->"])
    ax.set_yticks([])
    for l in ax.lines:
        l.set_color("#444444")
        l.set_linewidth(2)
        l.set_zorder(10)
    for s in ax.spines:
        ax.spines[s].set_color("#999999")
        ax.spines[s].set_alpha(0.6)
        ax.spines[s].set_linewidth(1)
    ax.spines["left"].set_position("center")
    ax.xaxis.set_label_coords(1.0, -0.025)
    ax.yaxis.set_label_coords(0.47, .9)
    return ax

f,ax = plt.subplots()
ax.plot(ht, F(ht,rho,kappa))
ax.set_xlabel(r"$h\tau$", fontdict={"size":14})
ax.set_ylabel(r"$F_t$", fontdict={"size":14}, rotation=0)
ax.set_title(r"$\delta$", size=20)
ax.arrow(0.8,0.5,.5,0, transform=ax.transAxes, length_includes_head=True, color="#444444", clip_on=False)
ax.annotate("", xy=(2, .2), xytext=(0.1,0.2),
            arrowprops=dict(arrowstyle="->", color="#444444", lw=1.5))
ilustration(ax)
```

```python
def Frhoseq(n=5, kap=0.15):
    f,ax = plt.subplots(1, n, figsize=(10,4), sharey="all")
    rho = np.linspace(0.001, 0.9, n)
    ht = np.arange(-5,5,0.1)
    for i in xrange(n):
        ax[i].plot(ht, F(ht,rho[i], kap))
        ilustration(ax[i])
        ax[i].set_xlabel(r"$\frac{\tau\,\overline{h}}{\gamma}$", size=18)
        ax[i].set_ylabel(r"$F_%d$"%i, size=14, rotation=0)
        #ax[i].set_title(r"$D_%d$"%i, size=14)
        ax[i].yaxis.label.set_position((0.4,0.9))
        ax[i].set_ylim((0,1))
        ax[i].text(.6, .6, r"$\rho_%d=%.1f$"%(i,rho[i]), transform=ax[i].transAxes,size=14),
        if i != n-1:
            ax[i].arrow(1.1,0.5,.25,0, transform=ax[i].transAxes,
                        length_includes_head=True, color="#444444", clip_on=False)
            ax[i].text(1.12,0.57, r"$+y_%d$"%(i+1), transform=ax[i].transAxes,size=14)
        f.tight_layout(pad=4)
    return f, ax

f,ax = Frhoseq(n=4)
#f.savefig(path+"modulation-function-rho.png", format="png")
```

```python
def Fkapseq(n=5, rho=0.5):
    f,ax = plt.subplots(figsize=(10,6))
    eps = np.linspace(0.0, 0.45, n)
    ht = np.arange(-3,3,0.1)
    g = np.sqrt(1-rho*rho)/rho
    styles = [":", "-.", "--", "-"]
    colors = ["#444444","#666666", "#888888", "#aaaaaa"]
    for i in xrange(n):
        ax.plot(ht, F(ht,rho, eps[i]),label=r"$\varepsilon=%.2f$"%eps[i], ls=styles[i], c=colors[i])
        ax.set_xlabel(r"$\frac{\overline{h}\tau}{\gamma}$", size=18)
        ax.set_ylabel(r"$F$", size=14, rotation=0)
        #ax[i].set_title(r"$D_%d$"%i, size=14)
        #x.yaxis.label.set_position((0.4,0.9))
        # = ax.get_ylim()[1]
        #x.set_ylim((0,y))
        #x.set_xlim((-3,3))
        #x.text(.6, .6, r"$\eta=%.1f$"%kap[i], transform=ax[i].transAxes,size=14))
    ilustration(ax)
    sns.despine(ax=ax)
    ax.set_xticks([0])
    ax.set_yticks([])
    ax.spines["left"].set_position("center")
    ax.set_xlabel(r"$\frac{\tau\,\overline{h}}{\gamma}$", size=18)
    ax.xaxis.set_label_coords(1.0, -0.025)
    ax.yaxis.set_label_coords(0.47, .95)
    ax.legend(loc=(0.8,0.6), prop={"size":18})
    return f, ax

f,ax = Fkapseq(n=4, rho=0.5)
f.savefig(path+"modulation-function-eta.png", format="png")
```

```python
def Fmf(x, rho, ns, x0):
    A = rho*(1+np.sign(x))/2
    B = (1-np.sign(x+(1-2*ns)*(x0+1e-3)))/2
    return 1 - A - B

def Vmf(x, rho, ns, x0):
    A = (1-rho)*x/2
    B = rho*np.abs(x)/2
    C = np.abs(x+(1-2*ns)*(x0+1e-3))/2
    return -A + B - C

x = np.linspace(-3,3,500)
x0 = x.max()
rho = 0.6
ns = 0.2
f, (ax1,ax2) = plt.subplots(nrows=2, ncols=1, figsize=(10,8))
l1, = ax1.plot(x, Fmf(x,rho,ns,x0), "--")
l2, = ax2.plot(x, Vmf(x,rho,ns,x0))
ax = (ax1,ax2)
for axes in ax:
    ilustration(axes)
    axes.set_xlabel(r"$z$", size=20)

ax1.set_ylim(top=1.1)
ax1.set_yticks([1-rho,1.05])
ax1.set_xticks([-x0*(1-2*ns), 0])
ax1.set_xticklabels([r"$-(1-2\varepsilon)z_0$", r"$0$"], size=15)
ax1.set_yticklabels([r"$1-\rho$",r"$1$"], size=15)
ax2.set_xticklabels([r"$0$"], size=15)
#(l1,l2),(r"$F_{MF}$",r"$V_{MF}$"), loc=(1.05,0), frameon=False
f.legend((l1,l2),(r"$F_{MF}$",r"$V_{MF}$"), loc=(0.8,0.8), frameon=False, prop={"size":20})
f.tight_layout(pad=3)
f.savefig(path+"mean-field.png", format="png")
```
