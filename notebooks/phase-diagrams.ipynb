```python
@np.vectorize
def fix_bug(x):
    return complex(x).real
```

```python
sim = "simulation-test-new-2/"
path = base_path + sim

statistics = pd.read_csv(path+"statistics.csv")
cumulative = pd.read_csv(path+"cumulative.csv")
social_network = np.load(path+"social-network.npy")
state = np.load(path+"state.npy")
zeitgeist = np.load(path+"zeitgeist.npy")

print(statistics.keys())
print(social_network.shape)
delta = statistics["delta"].reshape((6,100))
beta = statistics["beta"].reshape((6,100))
gamma = statistics["gamma"].reshape((6,100))
m = statistics["m_1"].reshape((6,100))
v = statistics["v_1"].reshape((6,100))
R_max = statistics["R_max"].reshape((6,100))
R_mean = statistics["R_mean"].reshape((6,100))
A_max = statistics["A_max"].reshape((6,100))
A_mean = statistics["A_mean"].reshape((6,100))
social_network = social_network.reshape((6,100,64,64))

c = ['r','c','b','y','g','m']
s = ['-', '--', '.--', '*--', 'v--', 'o--']

# -- fixing a strange bug -- 
m = fix_bug(m)
delta = fix_bug(delta)
beta = fix_bug(beta)
social_network = fix_bug(social_network)
R_mean = fix_bug(R_mean)
R_max = fix_bug(R_max)


# -- m x beta --
f = pl.figure(figsize=(9,7))
ax = f.add_subplot(111)
for i in xrange(6):
    img = ax.plot(beta[i], np.abs(m[i]), 
                  '%s%s'%(c[i],s[i]), label=r"$\delta = %.1f$"%delta[i,0])
ax.set_xlabel(r"$\beta$")
ax.set_ylabel(r"$|m|$")
ax.xaxis.label.set_fontsize(30)
ax.yaxis.label.set_fontsize(30)
ax.legend(loc=(1.1,0.05), prop={"size":20},frameon=False)
#txt1 = ax.text(5, .8, r"$\delta=%.1f$"%delta[i,3])
#txt.set_fontsize(15)
f.tight_layout()


# -- R x beta --
f2 = pl.figure(figsize=(9,7))
ax2 = f2.add_subplot(111)
for i in xrange(6):
    img2 = ax2.plot(beta[i], R_mean[i]/R_max[i], 
                  '%s%s'%(c[i],s[i]), label=r"$\delta = %.1f$"%delta[i,0])
ax2.set_xlabel(r"$\beta$")
ax2.set_ylabel(r"$\frac{\left<R\right>}{R_{max}}$")
ax2.xaxis.label.set_fontsize(30)
ax2.yaxis.label.set_fontsize(30)
ax2.legend(loc=(1.1,0.05), prop={"size":20},frameon=False)
#txt2 = ax2.text(5, .8, r"$\delta=%.1f$"%delta[i,3])
#txt2.set_fontsize(15)
f2.tight_layout()

# -- network --
f3 = pl.figure(figsize=(9,7))
ax3 = f3.add_subplot(111)
d, b = 2, 24
img3 = ax3.imshow(social_network[d,b,:,:])
f3.colorbar(img3)
```
