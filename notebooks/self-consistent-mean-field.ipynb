```python
from __future__ import division, print_function
import numpy as np
from scipy.integrate import quad
from scipy.optimize import fixed_point
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from IPython.display import display
from IPython.html import widgets
sns.set_style("white")
sns.set_style({"font.family":"Ubuntu", "axes.grid": False,
               "grid.color": "0.85", "font.size":20})
sns.set_context("notebook", {"lines.linewidth": 2.0,
                             "figure.figsize":np.array([12,10])})
sns.set_palette("Set1")
%matplotlib inline
```

```python
def Psi(h, m, r, rho, tau, betaK):
    A = (1 - rho)/2
    B = -rho/2
    psi = (1 - h*h)*np.exp(betaK*(A*m*h + B*r*np.abs(h) + np.abs(tau-m*h)/2 + tau/2))
    return psi

def Z(*args):
    z = quad(Psi, -1, 1, args=args)[0]
    return z

def m(z, *args):
    f = lambda x: x*Psi(x, *args)
    m = quad(f, -1, 1)[0]/z
    return m

def r(z, *args):
    f = lambda x: np.abs(x)*Psi(x,*args)
    r = quad(f, -1, 1)[0]/z
    return r

def sce(m0, r0, rho, tau, betaK):
    n = 0
    converged = False
    M = m0
    R = r0
    delta = .1
    while (not converged) and n < 500:
        M0 = M
        R0 = R
        z = Z(M0, R0, rho, tau, betaK)
        M = (1 - delta)*M0 + delta*m(z, M0, R0, rho, tau, betaK)
        R = (1 - delta)*R0 + delta*r(z, M0, R0, rho, tau, betaK)
        if np.abs(M - M0)<1e-5 and np.abs(R -R0)<1e-5:
            converged = True
        elif n == 499:
            print("Failed to converge!")
        n += 1
    return M, R

def P(h, m, r, rho, tau, betaK):
    z = Z(m, r, rho, tau, betaK)
    return Psi(h, m, r, rho, tau, betaK)/z
```

```python
rho = .7
tau = -.2
betaK = 50
m0 = .01
r0 = .9
sce(m0, r0, rho, tau, betaK)
```

```python
def get_float_slider(list_, trim=2):
    x0, x = min(list_), max(list_)
    n = len(list_)-1
    if n != 0:
        dx = (x-x0)/n
    else:
        dx = 0.0
    s = [round(a, trim) for a in (x0, x, dx)]
    sld = widgets.FloatSliderWidget(min=s[0], max=s[1], 
                                    step=s[2], value=s[0])
    return sld

def plot_data(m0, r0, rho, epsilon):
    m_data = data['m'][m0][r0][rho][epsilon]
    r_data = data['r'][m0][r0][rho][epsilon]
    f = plt.figure(figsize=(10, 10))
    ax = f.add_subplot(1,1,1)
    m_data.plot(ax=ax, label=r"$m$")
    r_data.plot(ax=ax, label=r"$r$")
    ax.set_xlabel(r"$\beta$", fontdict={"size":30})
    ax.legend(loc=(1.05, 0), prop={"size":30})
    sns.despine(f)
    
def plt_wgt():
    prmt = ['m0', 'r0', 'rho', 'epsilon']
    vl = [get_float_slider(x.tolist()) for x in data.index.levels]
    args = dict(zip(prmt, vl))
    wgt = widgets.interactive(plot_data, **args)
    return wgt
```
