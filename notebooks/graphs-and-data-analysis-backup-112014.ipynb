```python
%matplotlib inline
from __future__ import division, print_function
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import json
import os
import sys
from nbUtil import *
from nbSpin import *
from IPython.html import widgets
from IPython.display import display
import seaborn as sns
import matplotlib as mpl
import brewer2mpl as brw
import math

sns.set_style("white")
#sns.set_style("ticks")
sns.set_style({"font.family":"Ubuntu", "axes.grid": False,
               "grid.color": "0.85"})
sns.set_context("notebook", {"lines.linewidth": 2.0, "figure.figsize":np.array([12,10])})
sns.set_palette("Set1")
#sns.plotting_context(),sns.axes_style()
#mpl.rc()
```

```python
# ----- utitlity functions ------
def get_slice(list_, trim=2):
    x0, x = min(list_), max(list_)
    n = len(list_)-1
    if n != 0:
        dx = (x-x0)/n
    else:
        dx = 0.0
    s = [round(a, trim) for a in (x0, x, dx)]
    return s

def make_slider(slice_, value):
    if value is 'min':
        value = slice_[0]
    elif value is 'max':
        value = slice_[1]
    min_, max_, step = slice_
    return widgets.FloatSliderWidget(min=min_, max=max_, step=step, value=value)

def get_bounds(level, value):
    if type(level) is not type([]):
        lvl_values = list(set(stat_panel.items.get_level_values(level).tolist()))
        lvl_slice = get_slice(lvl_values)
        return make_slider(lvl_slice, value)
    else:
        if len(level) != len(value):
            raise Exception("One value needed for each widget")
        B = []
        for i in xrange(len(level)):
            lvl_values = list(set(stat_panel.items.get_level_values(level[i]).tolist()))
            lvl_slice = get_slice(lvl_values)
            B.append(make_slider(lvl_slice, value[i]))
        return B
    
def get_data(path, directory, verbose=False):
    global config, pre_options, stat_panel, rep_panel, w_panel
    path = path+directory+'/'
    with open(path + "config.json") as file_:
        config = json.load(file_)
        
    D = config["D"]
    z = np.ones(D)
    z *= np.sqrt(D)/np.linalg.norm(z)
    stat_panel = get_panel(path + "statistics.csv")
    rep_panel = get_panel(path + "reputation.csv")
    w_panel = get_panel(path + "state.csv")
    op_list = stat_panel.minor_axis.values.tolist()
    cmaps = {"Green":plt.cm.Greens, 
             "Red":plt.cm.Reds,
             "Blue":plt.cm.Blues,
             "Orange":plt.cm.Oranges,
             "Purple":plt.cm.Purples,
             "binary":plt.cm.binary,
             "coolwarm":plt.cm.coolwarm}
    y_values = {"type":"dropdown", "args":{"value":op_list[0], "values":op_list}}
    eps_min = {"type":"float", "args":{"level":"epsilon", "value":"min"}}
    eps_max = {"type":"float", "args":{"level":"epsilon", "value":"max"}}
    rho_min = {"type":"float", "args":{"level":"rho", "value":"min"}}
    rho_max = {"type":"float", "args":{"level":"rho", "value":"max"}}
    beta_max = {"type":"float", "args":{"level":"beta", "value":"max"}}
    beta_min = {"type":"float", "args":{"level":"beta", "value":"min"}}
    hist_qnt = {"type":"dropdown", "args":{"value":"overlap", "values":["overlap", "opinion"]}}
    hist_bins = {"type":"int", "args":{"min":1, "max":config["N"], "value":config["N"]}}
    mat_cmaps = {"type":"dropdown", "args":{"value":plt.cm.binary, "values":cmaps}}
    mat_qnt = {"type":"dropdown", "args":{"value":"reputation", "values":["overlap", "reputation"]}}
    corr_cmap1 = {"type":"dropdown", "args":{"value":plt.cm.Greens, "values":cmaps}}
    corr_cmap2 = {"type":"dropdown", "args":{"value":plt.cm.Purples, "values":cmaps}}
    x_eps_opt = {"op":y_values, "beta_1":beta_min, "beta_2":beta_max, "rho_1":rho_min, "rho_2":rho_max}
    x_beta_opt = {"op":y_values, "eps_1":eps_min, "eps_2":eps_max, "rho_1":rho_min, "rho_2":rho_max}
    hist_opt = {"quantity":hist_qnt, "eps_1":eps_min, "eps_2":eps_max, "rho_1":rho_min,
                "rho_2":rho_max, "beta":beta_max, "bins":hist_bins}
    mat_opt = {"quantity":mat_qnt, "eps":eps_min, "rho_1":rho_min, 
               "rho_2":rho_max, "beta":beta_max, "cmap":mat_cmaps}
    corr_opt = {"eps_1":eps_min, "eps_2":eps_max, "rho":rho_min, 
                "beta":beta_max, "cmap1":corr_cmap1, "cmap2":corr_cmap2}
    x_time_opt = {"y":y_values, "rho":rho_min, "beta":beta_max}
    pre_options = {"epsilon": (op_x_eps, x_eps_opt),
                   "beta": (op_x_beta, x_beta_opt),
                   "time": (op_x_time, x_time_opt),
                   "histograms": (plot_histograms, hist_opt),
                   "matrix": (plot_matrix, mat_opt),
                   "correlation": (plot_corr, corr_opt)}
    if verbose:
        print(json.dumps(config, indent=4, sort_keys=True))
        print(stat_panel)
        print(stat_panel.minor_axis)
        print(stat_panel.items.names)
    else:
        print(50*'\r')
           
# ------ plot functions ------
def op_x_eps(op="n_op", beta_1=0, beta_2=30, rho_1=.2, rho_2=.8):
    op_lbl = r"$%s$"%op
    if len(op.split('_')) > 1:
        p1, p2 = op.split('_')
        op_lbl = r"$%s_{%s}$"%(p1,p2)

    lbl = r"$\beta=%.1f$, $\rho=%.1f$"
    
    x = stat_panel.minor_xs(op)
    x = x.reorder_levels(["beta", "rho", "epsilon"],axis=1)
    
    statistic = lambda v: np.abs(np.mean(v))
    x11 = x[beta_1][rho_1].apply(statistic)
    x12 = x[beta_1][rho_2].apply(statistic)
    x21 = x[beta_2][rho_1].apply(statistic)
    x22 = x[beta_2][rho_2].apply(statistic)

    fg = plt.figure(figsize=(10,16))
    ax1 = fg.add_subplot(211)
    ax2 = fg.add_subplot(212)
    
    x11.plot(label=lbl%(beta_1, rho_1), ax=ax1)
    x12.plot(label=lbl%(beta_1, rho_2), ax=ax1)
    x21.plot(label=lbl%(beta_2, rho_1), ax=ax2)
    x22.plot(label=lbl%(beta_2, rho_2), ax=ax2)
    
    sns.despine(fg)
    ax1.set_xlabel(r"$\epsilon$", fontdict={"size":30})
    ax2.set_xlabel(r"$\epsilon$", fontdict={"size":30})
    ax1.set_ylabel(op_lbl, fontdict={"size":30})
    ax2.set_ylabel(op_lbl, fontdict={"size":30})
    ax1.legend(loc=(1.01, 0), prop={"size":20})
    ax2.legend(loc=(1.01, 0), prop={"size":20})
    ax1.tick_params(axis='both', which='major', labelsize=15)
    ax2.tick_params(axis='both', which='major', labelsize=15)
    fg.tight_layout()
    
def op_x_beta(op="m", eps_1=.0, eps_2=30, rho_1=.2, rho_2=.8):
    op_lbl = r"$%s$"%op
    if len(op.split('_')) > 1:
        p1, p2 = op.split('_')
        op_lbl = r"$%s_{%s}$"%(p1,p2)

    lbl = r"$\epsilon=%.1f$, $\rho=%.1f$"
    
    x = stat_panel.minor_xs(op)
    
    f = lambda x: np.abs(np.mean(x))
    
    x11 = x[eps_1][rho_1].apply(f)
    x12 = x[eps_1][rho_2].apply(f)
    x21 = x[eps_2][rho_1].apply(f)
    x22 = x[eps_2][rho_2].apply(f)

    fg = plt.figure(figsize=(10,16))
    ax1 = fg.add_subplot(211)
    ax2 = fg.add_subplot(212)
    
    x11.plot(label=lbl%(eps_1, rho_1), ax=ax1)
    x12.plot(label=lbl%(eps_1, rho_2), ax=ax1)
    x21.plot(label=lbl%(eps_2, rho_1), ax=ax2)
    x22.plot(label=lbl%(eps_2, rho_2), ax=ax2)
    
    sns.despine(fg)
    ax1.set_xlabel(r"$\beta$", fontdict={"size":30})
    ax2.set_xlabel(r"$\beta$", fontdict={"size":30})
    ax1.set_ylabel(op_lbl, fontdict={"size":30})
    ax2.set_ylabel(op_lbl, fontdict={"size":30})
    ax1.legend(loc=(1.01, 0), prop={"size":20})
    ax2.legend(loc=(1.01, 0), prop={"size":20})
    ax1.tick_params(axis='both', which='major', labelsize=15)
    ax2.tick_params(axis='both', which='major', labelsize=15)
    fg.tight_layout()

def plot_histograms(quantity="overlap",eps_1=.0, eps_2=.4, rho_1=.2, rho_2=.8, beta=0, bins=100):
    
    eps = [eps_1, eps_2]
    rho = [rho_1, rho_2]
    D = config["D"]
    w_list = [{(e,r):w_panel[e, r, beta].values} for e in eps for r in rho]
    
    x_dict = {"overlap": (map(lambda x: x.values()[0], w_list), r"$\rho_{ij}$"),
              "opinion": (4*[z],r"$h_i$")}
    x_def = x_dict[quantity]
    x_list = []
    x_lbl = x_def[1]
    for i in xrange(4):
        w = w_list[i]
        x = x_def[0][i]
        k, v = w.items()[0]
        x = x.dot(v.T)/D
        x_list.append({k:x})
            
    with sns.axes_style("ticks"):
        sns.set_style({"font.family":"Ubuntu", "axes.grid": False,
               "grid.color": "0.85"})
        fg = plt.figure(figsize=(12,10))
        ax_list = [fg.add_subplot(2,2,i) for i in xrange(1,5)]
        color_cycle = ax_list[0]._get_lines.color_cycle
        colors = [color_cycle.next(), color_cycle.next()]
        color_map = dict(zip(rho, colors))

        for i in xrange(4):
            k, v = x_list[i].items()[0]
            ax = ax_list[i]
            hist = ax.hist(v.ravel(), bins=bins, color=color_map[k[1]])
            ax.set_xlim(-1,1)

        eps_str = r"$\epsilon=%.2f$"
        rho_str = r"$\rho=%.2f$"
        ax_list[0].text(.4, 1.05, rho_str%rho_1, transform=ax_list[0].transAxes, fontdict={"size":30})
        ax_list[1].text(.4, 1.05, rho_str%rho_2, transform=ax_list[1].transAxes, fontdict={"size":30})
        ax_list[0].text(-.45, .45, eps_str%eps_1, transform=ax_list[0].transAxes, fontdict={"size":30})
        ax_list[2].text(-.45, .45, eps_str%eps_2, transform=ax_list[2].transAxes, fontdict={"size":30})
        ax_list[2].text(.45, -.2, x_lbl, transform=ax_list[2].transAxes, fontdict={"size":30})
        ax_list[3].text(.45, -.2, x_lbl, transform=ax_list[3].transAxes, fontdict={"size":30})
        sns.offset_spines()
        sns.despine(fg, trim=True)
        fg.tight_layout() 
        
    sns.axes_style({"xtick.major.size":"0", "ytick.major.size":"0",
                    "xtick.minor.size":"0", "ytick.minor.size":"0"})
    
    
def plot_matrix(quantity="reputation", eps=0.2, rho_1=.2, rho_2=.8, beta=9, cmap=plt.cm.binary):
    
    x_dict = {"reputation": rep_panel, "overlap": w_panel}
    x = x_dict[quantity]
    
    x1 = x[eps, rho_1, beta].values
    x2 = x[eps, rho_2, beta].values
    
    if quantity is "overlap":
        x1 = x1.dot(x1.T)
        x2 = x2.dot(x2.T)
        
    spin_x1, _ = spin(x1, s=2, m=100, n=100)
    spin_x2, _ = spin(x2, s=2, m=100, n=100)
    #spin_x1 = x1 - x1.T
    #spin_x1 *= spin_x1
    #spin_x2 = x2 - x2.T
    
    fg = plt.figure(figsize=(12,10))
    ax1 = fg.add_subplot(121)
    ax2 = fg.add_subplot(122)
    ax1.set_xticks([])
    ax1.set_yticks([])
    ax2.set_xticks([])
    ax2.set_yticks([])
    img1 = ax1.imshow(spin_x1, cmap=cmap, interpolation="nearest")#, vmin=0, vmax=1)
    img2 = ax2.imshow(spin_x2, cmap=cmap, interpolation="nearest")#, vmin=0, vmax=1)
    cb1 = plt.colorbar(img1, ax=ax1, fraction=.042, pad=.025, orientation="horizontal")
    cb2 = plt.colorbar(img2, ax=ax2, fraction=.042, pad=.025, orientation="horizontal")
    info = "$\\epsilon=%.1f$\n$\\rho=%.1f$\n$\\beta=%.1f$"
    ax1.text(-.275, .0, info%(eps, rho_1, beta), transform=ax1.transAxes, fontdict={"size":30})
    ax2.text(1.05, .0, info%(eps, rho_2, beta), transform=ax2.transAxes, fontdict={"size":30})
    sns.despine(fg, left=True, bottom=True)
    fg.tight_layout(w_pad=2)
    
def plot_corr(eps_1=.0, eps_2=.4, rho=.5, beta=9, cmap1=plt.cm.Greens, cmap2=plt.cm.Purples):
    w1 = w_panel[eps_1, rho, beta].values
    w2 = w_panel[eps_2, rho, beta].values
    r1 = rep_panel[eps_1, rho, beta].values.ravel()
    r2 = rep_panel[eps_2, rho, beta].values.ravel()
    rho1 = w1.dot(w1.T).ravel()/config["D"]
    rho2 = w2.dot(w2.T).ravel()/config["D"]
    
    fg = plt.figure(figsize=(10,16))
    ax1 = fg.add_subplot(211)
    ax2 = fg.add_subplot(212)

    x1 = r1+0.05*np.random.randn(r1.shape[0])
    y1 = rho1
    x2 = r2+0.05*np.random.randn(r2.shape[0])
    y2 = rho2
    
    lgt_gry = (.9,.9,.9)
    fill = 'w'
    #fill = (.1,.1,.1)
    cmap1.set_under(fill,.8)
    cmap2.set_under(fill,.8)
    img1 = ax1.hexbin(x1, y1, gridsize=40, cmap=cmap1, edgecolors=lgt_gry,norm=mpl.colors.LogNorm(vmin=1.01))
    img2 = ax2.hexbin(x2, y2, gridsize=40, cmap=cmap2, edgecolors=lgt_gry, norm=mpl.colors.LogNorm(vmin=1.01))
    info = "$\\epsilon=%.2f$\n$\\rho=%.2f$\n$\\beta=%.2f$"
    ax1.text(1.05, .0, info%(eps_1, rho, beta), transform=ax1.transAxes, fontdict={"size":30})
    ax2.text(1.05, .0, info%(eps_2, rho, beta), transform=ax2.transAxes, fontdict={"size":30})
    ax1.set_xlabel(r"$r_{ij}$", fontdict={'size':30})
    ax2.set_xlabel(r"$r_{ij}$", fontdict={'size':30})
    ax1.set_ylabel(r"$\rho_{ij}$", fontdict={'size':30})
    ax2.set_ylabel(r"$\rho_{ij}$", fontdict={'size':30})
    sns.despine()
    
def op_x_time(y="n_op", rho=.5, beta=9):
    eps_lvl = np.asarray(stat_panel.items.levels[0])
    n = eps_lvl.shape[0]
    stp = int(math.floor((n-2)/3))
    diff = n - (3*stp + 2)
    idx = [0, stp, 2*stp, 3*stp + diff, n-1]
    eps_lvl = eps_lvl[idx]
    for eps in eps_lvl:
        data = stat_panel.minor_xs(y)[eps,rho,beta]
        data.plot(label=r"$\epsilon=%.2f$"%eps)
    ax = plt.gca()
    y_lbl = r"$%s$"%y
    if len(y.split('_')) > 1:
        p1, p2 = y.split('_')
        y_lbl = r"$%s_{%s}$"%(p1,p2)

    info = "$\\rho=%.1f$\n$\\beta=%.1f$"
    ax.text(1.05, .45, info%(rho, beta), transform=ax.transAxes, fontdict={"size":30})
    ax.set_ylabel(y_lbl, fontdict={'size':30})
    ax.set_xlabel(r"$t$", fontdict={'size':30})
    ax.legend(loc=(1.05,0), prop={"size":20})
    sns.despine()

# ----- widget creators -----

def make_get_data_wgt(path="/media/felippe/backup/simulation-data/",
                      grep=None):
    config = None
    pre_options = None
    z = None
    stat_panel = None
    rep_panel = None
    w_panel = None
    all_dirs = os.listdir(path)
    valid = all_dirs
    if grep:
        valid = [x for x in all_dirs if grep in x]
    dir_list_wg = widgets.SelectWidget(values=valid,
                                       description="directory")
    verbos_wgt = widgets.CheckboxWidget(description="verbose", value=False)
    wgt = widgets.interactive(get_data, path=widgets.fixed(path),
                              directory=dir_list_wg, verbose=verbos_wgt)
    
    
    return wgt, pre_options, config, z, stat_panel, rep_panel, w_panel
        
def make_wgt(func, option):
    opt_action = {"int": widgets.IntSliderWidget,
                  "float": get_bounds,
                  "dropdown": widgets.DropdownWidget}
    args = {}
    for k,v in option.items():
        type_ = v["type"]
        params = v["args"]
        args[k] = opt_action[type_](**params)
    wgt = widgets.interactive(func, **args)
    return wgt

# ----- test -----
def op_x_cp(order, y, cp2_1, cp2_2, cp3_1, cp3_2):
    x, cp2, cp3 = order
    y_lbl = r"$%s$"%y
    if len(y.split('_')) > 1:
        p1, p2 = y.split('_')
        y_lbl = r"$%s_{%s}$"%(p1,p2)
    x_lbl = r"$\%s=%.2f$, $\%s=%.2f$"
    
    data = stat_panel.minor_xs(y)
    data = data.reorder_levels([cp2, cp3, x],axis=1)
    points = [(v1, v2) for v1 in [cp2_1, cp2_2] for v2 in [cp3_1, cp3_2]]
    pairs = [[(cp2_1, cp3_1), (cp2_1, cp3_2)], [(cp2_2, cp3_1), (cp2_2, cp3_2)]]
    
    #statistic = lambda v: np.abs(np.mean(v))
    statistic = np.mean
    table = {}
    for p in points:
        ax_idx = -1
        if p in pairs[0]:
            ax_idx = 0
        elif p in pairs[1]:
            ax_idx = 1
        table[p] = {"data":data[p[0]][p[1]].apply(statistic), "ax":ax_idx}

    fg, ax = plt.subplots(figsize=(10,16), nrows=2, ncols=1)
    for k, v in table.items():
        v["data"].plot(ax=ax[v["ax"]], label=x_lbl%(cp2, k[0], cp3, k[1]))
      
    sns.despine(fg)
    for axes in ax:
        axes.set_xlabel(r"$\%s$"%x, fontdict={"size":30})
        axes.set_ylabel(y_lbl, fontdict={"size":30})
        axes.legend(loc=(1.01, 0), prop={"size":20})
        axes.tick_params(axis='both', which='major', labelsize=15)
        
    fg.tight_layout()
```
