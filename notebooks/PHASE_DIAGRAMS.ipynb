```python
from __future__ import division, print_function
import numpy as np
import pandas as pd
import pylab as pl
import json
import os
%matplotlib inline
```

```python
data_dir = "/media/backup/simulation-data/ms-project/data/test-42/"
os.chdir(data_dir)

config = json.load(open("config.json","r"))
print(config["readme"], config.keys())
statistics = pd.read_csv("statistics.csv")
social_network = np.load("social-network.npy")
state = np.load("state.npy")
zeitgeist = np.load("zeitgeist.npy")

N = config["society"]["N"]
D = config["society"]["D"]
bbeta = config["bounds"]["beta"]
bdelta = config["bounds"]["delta"]
bgamma = config["bounds"]["gamma"]
nbeta = np.arange(*bbeta).shape[0]
ndelta = np.arange(*bdelta).shape[0]
ngamma = np.arange(*bgamma).shape[0]

# index order: gamma,delta,beta
new_shape = (ngamma, ndelta, nbeta)
beta = statistics["beta"].reshape(new_shape)
delta = statistics["delta"].reshape(new_shape)
gamma = statistics["gamma"].reshape(new_shape)
m = statistics["m"].reshape(new_shape)
n = statistics["n"].reshape(new_shape)
r = statistics["r"].reshape(new_shape)
v = statistics["v"].reshape(new_shape)
stag_m = statistics["stag_m"].reshape(new_shape)
M = statistics["M"].reshape(new_shape)
R_max = statistics["R_max"].reshape(new_shape)
R_mean = statistics["R_mean"].reshape(new_shape)
R_var = statistics["R_mean"].reshape(new_shape)


sn = social_network.reshape((ngamma,ndelta,nbeta,N,N))
x = state.reshape((ngamma,ndelta,nbeta,N,D))
all_z = zeitgeist.reshape((ngamma,ndelta,nbeta,D))

c = ['r','c','b','y','g','m']
s = ['-', '--', '.--', '*--', 'v--', 'o--']
figshape = (11,9)

i = 4 # delta index
b1 = 10  # beta indices
b2 = 96


# --- magnetization ---
f1 = pl.figure(figsize=figshape)
ax1 = f1.add_subplot(111)
img =  ax1.plot(beta[0,i,:], np.abs(m[0,i,:]), 'r-o', label="$|m|$")
img =  ax1.plot(beta[0,i,:], stag_m[0,i,:], 'c-o', label="$m_s$")
#img =  ax1.plot(beta[0,i,:], np.abs(n[0,i,:]), 'k-o', label="$n$", alpha=.5)
#img =  ax1.plot(beta[0,i,:], r[0,i,:], 'y-o', label="$r$", alpha=.5)
ax1.set_xlabel(r"$\beta$")
ax1.xaxis.label.set_fontsize(30)
ax1.yaxis.label.set_fontsize(30)
ax1.legend(loc=(1.05,0), frameon=False,  prop={"size":30})
ax1.set_title("$\delta=%.1f,\;\gamma=-%.1f$"%(delta[0,i,0],gamma[0,i,0]), 
              fontdict={"fontsize":30})

f2 = pl.figure(figsize=figshape)
ax2 = f2.add_subplot(111)
img =  ax2.plot(beta[2,i,:], np.abs(m[2,i,:]), 'r-o', label="$|m|$")
img =  ax2.plot(beta[2,i,:], stag_m[2,i,:], 'c-o', label="$m_s$")
#img =  ax2.plot(beta[2,i,:], np.abs(n[2,i,:]), 'k-o', label="$n$", alpha=.5)
#img =  ax2.plot(beta[2,i,:], r[2,i,:], 'y-o', label="$r$", alpha=.5)
ax2.set_xlabel(r"$\beta$")
ax2.xaxis.label.set_fontsize(30)
ax2.yaxis.label.set_fontsize(30)
ax2.legend(loc=(1.05,0), frameon=False,  prop={"size":30})
ax2.set_title("$\delta=%.1f,\;\gamma=-%.1f$"%(delta[2,i,0],gamma[2,i,0]), 
              fontdict={"fontsize":30})

# --- reputation ---
f3 = pl.figure(figsize=figshape)
ax3 = f3.add_subplot(111)
img =  ax3.plot(beta[0,i,:], R_mean[0,i,:]/R_max[0,i,:], 
               'r-o',#%s%s"%(c[i],s[i]),
               label="$\delta=%.1f,\;\gamma=-%.1f$"%(delta[0,i,0],gamma[0,i,0]))
img =  ax3.plot(beta[2,i,:], R_mean[2,i,:]/R_max[2,i,:], 
               'c-o',#%s%s"%(c[i],s[i]),
               label="$\delta=%.1f,\;\gamma=-%.1f$"%(delta[2,i,0],gamma[2,i,0]))
ax3.set_xlabel(r"$\beta$")
ax3.set_ylabel(r"$\frac{\left<R\right>}{R_{max}}$")
ax3.xaxis.label.set_fontsize(30)
ax3.yaxis.label.set_fontsize(30)
ax3.legend(loc=(1.05,0), frameon=False,  prop={"size":20})

# --- opinion histograms ---
s1 = x[0,i,b1,:,:]
z1 = all_z[0,i,b1,:]
h1 = z1.dot(s1.T)
s2 = x[-1,i,b1,:,:]
z2 = all_z[-1,i,b1,:]
h2 = z2.dot(s2.T)

f4 = pl.figure(figsize=(figshape))
ax4 = f4.add_subplot(111)
img = ax4.hist(h1, bins=10, normed=False, rwidth=.2, color="red", alpha=.5,
               label="$\gamma=-%.1f$"%gamma[0,i,0])
img = ax4.hist(h2, bins=10, normed=False, rwidth=.2, color="cyan", alpha=.5,
               label="$\gamma=-%.1f$"%gamma[-1,i,0])
ax4.set_xlabel("$h$")
ax4.xaxis.label.set_fontsize(30)
ax4.set_title("$\delta=%.1f,\; beta=%.1f$"%(delta[0,i,0],beta[0,i,b1]),
              fontdict={"fontsize":30})
ax4.legend(loc=(1.02,0), prop={"size":20}, frameon=False)

s3 = x[0,i,b2,:,:]
z3 = all_z[0,i,b2,:]
h3 = z3.dot(s3.T)
s4 = x[-1,i,b2,:,:]
z4 = all_z[-1,i,b2,:]
h4 = z4.dot(s4.T)

f5 = pl.figure(figsize=figshape)
ax5 = f5.add_subplot(111)
img = ax5.hist(h3, bins=10, normed=False, rwidth=.2, color="red", alpha=.5,
               label="$\gamma=-%.1f$"%gamma[0,i,0])
img = ax5.hist(h4, bins=10, normed=False, rwidth=.2, color="cyan", alpha=.5,
               label="$\gamma=-%.1f$"%gamma[-1,i,0])
ax5.set_xlabel("$h$")
ax5.xaxis.label.set_fontsize(30)
ax5.set_title("$\delta=%.1f,\;beta=%.1f$"%(delta[0,i,0],beta[0,i,b2]), 
              fontdict={"fontsize":30})
ax5.legend(loc=(1.02,0), prop={"size":20}, frameon=False)
```
