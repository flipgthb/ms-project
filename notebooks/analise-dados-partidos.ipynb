```python
def lerdados(mandato):
    patterns = (
        ("codigo", re.compile(r"CD\w+ \w+", re.UNICODE)), # procura por CDx... y... onde x e y são sequências de letras ou números
        ("deputado", re.compile(r"(\w+)?(-)?\w+(\.)? (\w+)?", re.UNICODE)), # procura por até 3 palavras, onde a primeira admite hífen e ponto
        ("voto", re.compile(ur"(Sim|Nao|Abstencao|Presente|Art. \w+|<(-)+>|Obstrucao|Branco)", re.UNICODE)), 
        ("partido", re.compile(r"(\w{2,5}|S.Part.)", re.UNICODE)),
        ("estado", re.compile(ur"\w+", re.UNICODE))
    )
    votos = pd.DataFrame()
    for listavotos in glob.glob(mandato+"LVCD*"):
        _, pastavotos = listavotos.split("/")
        controle = glob.glob(mandato+"HECD"+pastavotos[4:]+"*")
    
        if len(controle) > 1:
            raise Exception("algo errado: final repetido")
        else:
            controle = controle[0]
        
        _,sumario = controle.split("/")
        with open(controle) as entradacontrole:
            lines = entradacontrole.readlines()
            dia, pl = lines[2][:-2], lines[-1][:-2]
        with codecs.open(listavotos, encoding="latin-1") as entradavotos:
            lines = entradavotos.readlines()
            for i in xrange(len(lines)):
                l = lines[i]
                l0 = l
                l = unicodedata.normalize('NFKD', l).encode('ascii', 'ignore')
                lines[i] = l
                serie = pd.Series()
                for info, pat in patterns:
                    m = pat.search(l)
                    try:
                        serie[info] = l[m.start():m.end()]
                        l = l[m.end():]
                    except:
                        raise Exception("procurando %s em %s achou nada na linha %s"%(info,listavotos,l0))
                serie["dia"] = dia
                serie["pl"] = pl
                votos = votos.append(serie, ignore_index=True)
    return votos
```

```python
governos = {}
mandatos = {"fhc1":"95-99", "fhc2":"99-03", "lula1":"03-07", "lula2":"07-11"}
for p, m in mandatos.items():
        governos[p] = {"dados":pd.read_csv("deputados%s.csv"%m, index_col=0, encoding="latin-1")}
        
for p, x in governos.items():
    v = x["dados"]
    pls = v["pl"].unique()
    valorvotos = v["voto"].unique()
    df = v.set_index("pl")
    df = df.loc[:,["voto", "partido"]]
    df = df.replace("Sim", 1)
    df = df.replace("Nao", -1)
    for y in valorvotos:
        df = df.replace(y, 0)
    pt = df.loc[df.partido == "PT", :]
    psdb = df.loc[df.partido == "PSDB", :]
    pmdb = df.loc[df.partido == "PMDB", :]
    mptpmdb = []
    mpsdbpmdb = []
    mptpt = []
    mpsdbpsdb = []
    mptpsdb = []
    for pl in pls:
        try:
            votospt = pt.loc[pl, "voto"].values
            votospsdb = psdb.loc[pl, "voto"].values
            votospmdb = pmdb.loc[pl, "voto"].values
            Cptpmdb = votospt*votospmdb[:,None]
            Cpsdbpmdb = votospsdb*votospmdb[:,None]
            Cptpt = votospt*votospt[:,None]
            Cpsdbpsdb = votospsdb*votospsdb[:,None]
            Cptpsdb = votospt*votospsdb[:,None]
        except:
            continue
        mptpmdb.append(Cptpmdb.mean())
        mpsdbpmdb.append(Cpsdbpmdb.mean())
        mptpt.append((Cptpt-np.diag(np.diag(Cptpt))).mean())
        mpsdbpsdb.append((Cpsdbpsdb-np.diag(np.diag(Cpsdbpsdb))).mean())
        mptpsdb.append(Cptpsdb.mean())
    #governos[p]["mptpmdb"] = np.array(mptpmdb)
    #governos[p]["mpsdbpmdb"] = np.array(mpsdbpmdb)
    #governos[p]["mptpt"] = np.array(mptpt)
    #governos[p]["mpsdbpsdb"] = np.array(mpsdbpsdb)
    #governos[p]["mptpsdb"] = np.array(mptpsdb)
    governos[p] = pd.DataFrame(data=np.vstack([mptpmdb, mpsdbpmdb, mptpt, mpsdbpsdb, mptpsdb]).T, 
                               columns="mptpmdb mpsdbpmdb mptpt mpsdbpsdb mptpsdb".split(" "))
    if np.any(np.isnan(governos[p].values)):
        print p, " got nan"
data = pd.Panel(governos)
#plt.plot(m31, "--r", alpha=0.8)
#plt.plot(m32, "-b", alpha=0.6)
```

```python
def resample(df,n):
    nrows, ncols = df.shape
    cols = df.columns
    rsix = np.random.randint(0, nrows, (n-1,ncols,nrows))
    sl = []
    for ix in rsix:
        #print ix.shape
        sdf = pd.DataFrame()
        for i in xrange(ix.shape[0]):
            c = cols[i]
            l = ix[i]
            v = df[c].values[l]
            sdf[c] = v
        sl.append(sdf)
    resample = [df] + sl
    return resample

foo = resample(data["fhc1"].dropna(axis=0), 10)
```
