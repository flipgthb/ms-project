```python
# ---- order parameters x tau at fixed beta values----

beta_low = 10
beta_high = 30
delta_low = 0.8
delta_high = 0.3
op = 'n_op'
op_text = "n_{op}"
#op = 'R'
#op_text = "R"
lbl_bldl = r"$%s$, $\delta=%.1f$, $\beta=%.1f$"%(op_text, 
                                                 delta_low, beta_low)
lbl_bldh = r"$%s$, $\delta=%.1f$, $\beta=%.1f$"%(op_text, 
                                                 delta_high, beta_low)
lbl_bhdl = r"$%s$, $\delta=%.1f$, $\beta=%.1f$"%(op_text, 
                                                 delta_low, beta_high)
lbl_bhdh = r"$%s$, $\delta=%.1f$, $\beta=%.1f$"%(op_text, 
                                                 delta_high, beta_high)
qr_str = "delta == %f and beta == %f"

f1 = plt.figure(figsize=(12,10))
ax1 = f1.add_subplot(111)
f2 = plt.figure(figsize=(12,10))
ax2 = f2.add_subplot(111)

df = stat_panel.to_frame()

x_bldl = df.T.query(qr_str%(delta_low, beta_low))
x_bldh = df.T.query(qr_str%(delta_high, beta_low))
x_bhdl = df.T.query(qr_str%(delta_low, beta_high))
x_bhdh = df.T.query(qr_str%(delta_high, beta_high))

x_bldl.index = x_bldl.index.levels[0]
x_bldh.index = x_bldh.index.levels[0]
x_bhdl.index = x_bhdl.index.levels[0]
x_bhdh.index = x_bhdh.index.levels[0]

x_bldl = x_bldl.T
x_bldh = x_bldh.T
x_bhdl = x_bhdl.T
x_bhdh = x_bhdh.T

op_bldl = x_bldl.query("minor == '%s'"%op).apply(np.mean)
op_bldh = x_bldh.query("minor == '%s'"%op).apply(np.mean)
op_bhdl = x_bhdl.query("minor == '%s'"%op).apply(np.mean)
op_bhdh = x_bhdh.query("minor == '%s'"%op).apply(np.mean)

op_bldl.plot(label=lbl_bldl, ax=ax1)
op_bldh.plot(label=lbl_bldh, ax=ax1)
op_bhdl.plot(label=lbl_bhdl, ax=ax2)
op_bhdh.plot(label=lbl_bhdh, ax=ax2)

f1.tight_layout()
f2.tight_layout()
ax1.legend(loc=(1.01,0), frameon=False)
ax2.legend(loc=(1.01,0), frameon=False)
ax1.set_xlabel(r"$\tau$", fontdict={"size":40})
ax2.set_xlabel(r"$\tau$", fontdict={"size":40})
ax1.set_ylabel(r"$%s$"%op_text, fontdict={"size":40})
ax2.set_ylabel(r"$%s$"%op_text, fontdict={"size":40})
remove_border(ax1)
remove_border(ax2)
```

```python
f = plt.figure(figsize=(12,10))

ax1 = f.add_subplot(221)
ax3 = f.add_subplot(223)
w1 = w_panel[.1,0.8,30].values
w2 = w_panel[.9,0.8,30].values
h1 = z.dot(w1.T)
h2 = z.dot(w2.T)
ax1.hist(h1, bins=30, label=r"$\gamma=-0.1$")
ax3.hist(h2, bins=30, label=r"$\gamma=-0.9$")
ax1.set_title(r"opinion histogram. $\delta=0.8$ and $\beta=30$")
ax1.legend(loc="best")
ax3.legend(loc="best")
remove_border(ax1)
remove_border(ax3)

ax2 = f.add_subplot(222)
ax4 = f.add_subplot(224)
ro1 = w1.dot(w1.T)
ro2 = w2.dot(w2.T)
ax2.hist(ro1.ravel(), bins=100, label=r"$\gamma=-0.1$")
ax4.hist(ro2.ravel(), bins=100, label=r"$\gamma=-0.9$")
ax2.set_xlim(-1,1)
ax2.set_title(r"overlap histogram $\delta=0.8$ and $\beta=30$")
ax2.legend(loc="best")
ax4.legend(loc="best")
remove_border(ax2)
remove_border(ax4)

f.tight_layout()
```
